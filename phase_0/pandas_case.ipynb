# Query Data

Query data digunakan untuk melihat data yang lebih presisi

```python
# df[conditions]
df_only_huskies= df[
    df["fran_id"] == "Huskies"
]
```

```python
df_only_huskies.head()
```

```python
df[
    (df["fran_id"] == "Huskies") & (df["pts"] >=60) & (df["opp_pts"] > 80)
]
```

```python
df[
    (df["fran_id"] == "Lakers") & (df["year_id"] >= 2011) & (df["pts"] >=90) 
]
```

```python
df.query('fran_id == "Lakers" & year_id == 1992') # harus beda kutip antara yang utama dan isi, df.query('')
```
